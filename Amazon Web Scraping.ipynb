## Let's Play with Beautisoup

```python
from bs4 import BeautifulSoup as bs
import requests
```

### Load amazon web page using request

```python
base_url = 'https://www.amazon.in'
```

```python
req = requests.get(base_url,headers={"User-Agent": "Requests"})
# convert into beautiful soup object

bsoup = bs(req.content)

prettified = bsoup.prettify()
print(prettified)
```

### search for specific product on amazon using modified url

```python
# in order to search asus laptops on amazon
search_query = '/s?k=asus+laptops'
req = requests.get(base_url+search_query,headers={"User-Agent": "Requests"})
bsoup = bs(req.content)
prettified = bsoup.prettify()

print(prettified)
```

```python
def print_arr(arr):
    for item in arr:
        print(item)
```

```python
def filter_product_name(card):
    product = card.find('h2')
    if product != None and product.find('a'):
        name = product.find('a').text
        return name
```

```python
product_cards = bsoup.find_all(class_ = "sg-col-inner")
products = []
for card in product_cards:
    product_name = filter_product_name(card)
    if product_name is not None:
        products.append(product_name)
print_arr(products)
print(type(products[0]))    

    
```

```python
def filter_by_comma(product):
    d = dict()
    a = product.split('(')
    key = a[0]
    values = []
    for index,val in enumerate(a):
        if index == 0 : continue
        else :
            val.replace(')','')
            values.append(val)
    d[key] = values
    return d
```

```python
def filter_product_name(product):
    a = product.split('(')
    name = a[0].replace("\n","")
    return name
```
